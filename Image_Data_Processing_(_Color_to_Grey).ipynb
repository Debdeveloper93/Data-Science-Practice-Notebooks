Resizing the images using pillow library

```python
from PIL import Image
```

```python
img = Image.open("/content/David-Goggins-workout-750x699.jpg")
img_resized = img.resize((200,200))
```

```python
print(img_resized)
```

```python
img_resized.save("image_resized.jpg")
```

```python
img_res = mpimg.imread("/content/image_resized.jpg")
img_res_plot = plt.imshow(img_res)
plt.show()
```

Convert RBG to greyscale

```python
import cv2
```

```python
img = cv2.imread("/content/David-Goggins-workout-750x699.jpg")
```
